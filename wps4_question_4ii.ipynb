```python
import numpy as np
import matplotlib.pyplot as plt
```

**WPS 4 Question 4**

The partition function over $j$ is given by:
\begin{equation}
Z_\text{rot} = \sum_{j=0}^\infty (2j+1)e^{-E(j)/kT} = \sum_{j=0}^\infty (2j+1)e^{-j(j+1)\epsilon/kT}.
\end{equation}

```python
def term_f(j, epsilon, k, T):
    return (2*j+1) * np.exp(-j*(j+1)*epsilon/(k*T))
```

```python
k = 8.62 * 10**(-5)
T = 300
epsilon = 2.4 * 10**(-4)
```

```python
approx_int = (k*T)/epsilon
```

```python
j = np.arange(0, 100 + 1, 1)
Z_rot = np.cumsum(term_f(j, epsilon, k, T))
```

```python
plt.plot(j, Z_rot)
plt.axhline(y=approx_int, color='r', linestyle='--', label=f'approx_int = {approx_int:.2f}')
plt.xlabel(r'$j$', fontsize=14)
plt.ylabel(r'$Z_\text{rot}$', fontsize=14)
plt.xlim(0, 100)
plt.legend()
plt.show()
```
