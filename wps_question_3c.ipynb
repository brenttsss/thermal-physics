```python
import numpy as np
import matplotlib.pyplot as plt
```

**WPS 4: Question 3(c)**

Given:
\begin{equation}
\tilde{p} = \frac{8\tilde{T}}{3\tilde{V}-1} - \frac{3}{\tilde{V}^2},
\end{equation}
and
\begin{equation}
\frac{G}{NkT_c} = -\tilde{T}\ln(3\tilde{V}-1) + \frac{\tilde{T}}{3\tilde{V}-1} - \frac{9}{4\tilde{V}},
\end{equation}
we want to plot $G/NkT$ versus $p$.


```python
def p(v, t):
    return ((8*t)/(3*v-1)) - (3/(v**2))

def g(v, t):
    return -t*np.log(3*v-1) + (t/(3*v-1)) - (9/(4*v))
```

```python
V = np.linspace(0.5, 6, 1000)
T = 0.83
```

```python
plt.plot(V, p(V,T))
plt.ylim(-0.4, 1.0)
plt.xlim(0, 6)
plt.xlabel(r'$\tilde{V}$')
plt.ylabel(r'$\tilde{p}$')
plt.show()
```

```python
plt.plot(p(V,T), g(V,T))
plt.xlabel(r'$\tilde{p}$')
plt.ylabel(r'$G/NkT$')
plt.show()
```
